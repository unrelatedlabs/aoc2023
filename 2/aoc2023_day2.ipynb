```python

available = {"red":12, "green":13, "blue":14}
```

```python
from collections import defaultdict

def parse_game(line):
    game, cubes = line.split(":")
    game_id = game.split(" ")[1]
    subgames = cubes.split(";")
    mins = defaultdict(int)
    is_valid = True 
    for subgame in subgames:
        for cube in subgame.split(","):
            # print(cube)
            num, color = cube.strip().split(" ")
            num = int(num)
            color = color.strip()
            if available[color] < num:
                is_valid = False
            mins[color] = max(mins[color], num)
    print(mins)
            
    return is_valid, int(game_id), mins["red"] * mins["green"] * mins["blue"]
       
```

```python
acc = 0 
acc2 = 0 
for line in data:
    ok, num, mins  = parse_game(line.strip())
    print(ok, num, mins)
    if ok:
        acc += num
    acc2 += mins

acc ,acc2
```
